```python
#jsonから抽出　　"brand_name","generic_name","active_ingredient","product_type","manufacturer_name","route","substance_name","indications_and_usage"]
import json
import pandas as pd

loaded_json_df = []
for l in range(1,11):#range(1,9)で0001から0008までを処理する
    json_locate = "../../../jupyter/FDA-drug-label/drug-label-0010/drug-label-" + "{0:04}".format(l) + "-of-0010.json"
    json_file = open(json_locate, "r", encoding="utf-8")
    loaded_json_file = json.load(json_file)

    loaded_json_df = pd.DataFrame(columns = 
                                  ["brand_name",
                                   "generic_name",
                                   "active_ingredient",
                                   "product_type",
                                   "manufacturer_name",
                                   "route",
                                   "substance_name",
                                   "package_ndc",
                                   "pharm_class_epc",
                                   "pharm_class_pe",
                                   "pharm_class_moa",
                                   "indications_and_usage",
                                   "marketing_status"
                                  ])
    json_to_excel = "../../../jupyter/FDA-drug-label/drug-label-0010/drug-label-" + "{0:04}".format(l) + "-of-0010.xlsx"

    for k in range(0,len(loaded_json_file["results"])):
        loaded_json_df = \
loaded_json_df.append(pd.Series(\
[loaded_json_file["results"][k]["openfda"].get("brand_name"),\
loaded_json_file["results"][k]["openfda"].get("generic_name"),\
loaded_json_file["results"][k].get("active_ingredient"),\
loaded_json_file["results"][k]["openfda"].get("product_type"),\
loaded_json_file["results"][k]["openfda"].get("manufacturer_name"),\
loaded_json_file["results"][k]["openfda"].get("route"),\
loaded_json_file["results"][k]["openfda"].get("substance_name"),\
loaded_json_file["results"][k]["openfda"].get("package_ndc"),\
loaded_json_file["results"][k]["openfda"].get("pharm_class_epc"),\
loaded_json_file["results"][k]["openfda"].get("pharm_class_pe"),\
loaded_json_file["results"][k]["openfda"].get("pharm_class_moa"),\
loaded_json_file["results"][k].get("indications_and_usage"),\
loaded_json_file["results"][k].get("marketing_status")],\
index=loaded_json_df.columns),ignore_index=True)
    
    for j in range(0,len(loaded_json_df.columns)):
        for i in range(0,len(loaded_json_df.index)):
            try:
                loaded_json_df[loaded_json_df.columns[j]].iloc[i] = ", ".join(map(str,loaded_json_df[loaded_json_df.columns[j]].iloc[i][:]))
            except:
                loaded_json_df[loaded_json_df.columns[j]].iloc[i] = "NA"
            
    loaded_json_df.to_excel(json_to_excel,index=False)
```
